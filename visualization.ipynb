```python
import matplotlib.pyplot as plt
import torch

def visualize_csi(real_data, synthetic_data, real_labels, synthetic_labels, num_samples=5):
    fig, axs = plt.subplots(num_samples, 2, figsize=(10, num_samples * 3))
    fig.suptitle('Comparison of Real and GAN-Generated CSI Samples with Labels', fontsize=16)

    for i in range(num_samples):
        real_sample = real_data[i].cpu().numpy()
        synthetic_sample = synthetic_data[i].cpu().numpy()

        # Plot real sample with label
        axs[i, 0].imshow(real_sample, aspect='auto', cmap='viridis')
        axs[i, 0].set_title(f'Real CSI Sample - Label: {real_labels[i].item()}')
        axs[i, 0].axis('off')

        # Plot synthetic sample with label
        axs[i, 1].imshow(synthetic_sample, aspect='auto', cmap='viridis')
        axs[i, 1].set_title(f'Synthetic CSI Sample - Label: {synthetic_labels[i].item()}')
        axs[i, 1].axis('off')

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
    plt.savefig('real_fake_comparision')
```

```python
real_data = torch.load('data_mm_user1.pth')
real_labels = torch.load('labels_user1.pth')

synthetic_data = torch.load('cwgan_generated_samples\CWGAN_GP_data.pth')
synthetic_labels = torch.load('cwgan_generated_samples\CWGAN_GP_labels.pth')

visualize_csi(real_data, synthetic_data, real_labels, synthetic_labels)
```

```python
real_csi_data = torch.load("data_mm_user1.pth") 
synthetic_csi_data = torch.load("replica_synthetic_data.pth")

def plot_comparison(real_data, synthetic_data, num_samples=5):
    
    real_data = real_data.numpy()
    synthetic_data = synthetic_data.numpy()

    # Set up the plot
    fig, axes = plt.subplots(nrows=num_samples, ncols=2, figsize=(10, 2 * num_samples))

    for i in range(num_samples):
        # Plot real data
        axes[i, 0].imshow(real_data[i].reshape(50, 30), cmap='viridis') 
        axes[i, 0].set_title(f"Real Sample {i + 1}")
        axes[i, 0].axis('off')

        # Plot synthetic data
        axes[i, 1].imshow(synthetic_data[i].reshape(50, 30), cmap='viridis')  
        axes[i, 1].set_title(f"Synthetic Sample {i + 1}")
        axes[i, 1].axis('off')

    plt.tight_layout()
    plt.show()
    
    
plot_comparison(real_csi_data, synthetic_csi_data)
```
